#### Calculating the mean value for the "normalized-losses" coloumn

```python
df["bore"] = df["bore"].replace(np.NaN, avg_bore)
df["bore"]
```

```python
# Calculate mean for the stroke column 

avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"] = df["stroke"].replace(np.NaN, avg_stroke)
```

```python
# calcutaing mean for the horsepower cloumn

avg_horse_power = df['horsepower'].astype('float').mean(axis=0)
print("Average of Horsepower:", avg_horse_power)
```

```python
df["horsepower"] = df["horsepower"].replace(np.NaN, avg_horse_power)
```

```python
# calculating the mean for the peak rpm column

avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average of the peak-rpm:", avg_peak_rpm)
```

```python
df["peak-rpm"] = df["peak-rpm"].replace(np.NaN, avg_peak_rpm)
```

### Data Standardization 

We Usually collect data from different agencies in different formats. Data standardization is also a term for a particular type of data normalization where we substract the mean and dive by the standard deviation.

#### What is standardization

Strandardization is the process of transforming data into a common format, allowing the researcher to make a meaningful comparison.

Example 
Transform mpg to L/100km:

In the data set, fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we're developing an application in a country that acceps the fuel consumption with l/100km standard.

we need to apply data transformation to transform mpg to l/100km.

Formula for unit conversion: L/100km =235/mpg

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']

# Check for transformed data
df.head()
```

```python
# tranforming mpg tp l/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

#rename column name from highway-mpg to highway-l/100km
df.rename(columns={"highway-mpg":'highway-L/100km'},inplace = True)

#Check for transformed data
df.head()
```

### Data Normalization

##### Why Normalization?

Normalization is the process of transforming values of several variables into a similar range.
Typical normalization include:

   1. Scaling the variable so the variable average is 0
   2. scaling the variable so the variance is 1
   3. scaling the variable so the variable values range from 0 to 1

#### Example

To demonstrate the normalization, say to want to sacle the columns "length", "width", and "height".

Target - normalize theose variables so their values ranges from 0 and 1 

Approach - replace the original value by (original value)/(maximum value)

```python
# replace (original value) by (original value)/(maximum value)

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()

df[['length', 'width', 'height']].head()
```

### Binning

#### Why Binning

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.

#### Example 

In the data set, 'horsepoer' is a real valued variable rangeing from 48 to 288 and it has 59 unique values, Now we want the data for price difference between cars with high horsepower, medium and little horse power (3 types), we can reaarnage them into 3 bins to simplify analysis.

Use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins.

```python
#converting data into correct format 
df['horsepower'] = df['horsepower'].astype(int, copy=True)
```

```python
#Plot the histogram to see the horsepower of horsepower
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

#set x/yu lables and plot title
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')

plt.pyplot.show()
```

Find 3 bins of equal size bandwidth by using NUMPY'Slinespace(start_value, end_value, numbers generated function

sincine I want to include miinum value of the horsepower, set start_value = min(df["horsepower"]).

for max, set end_value = max(["horsepower"])

Since we're building 3 bins of equal length, I need 4 dividers, so numbers_generated = 4

Building a bin array with minimum value to a maximum value by using the bandwidth calculated above. The values will determine when one bin ends and another begins.

```python
bins = np.linspace(min(df['horsepower']),max(df['horsepower']), 4)
bins
```

```python
#set group names
group_names = ['Low', 'Medium', 'High']
```

```python
#applying the function 'cut' to determine what each value of df['horsepower'] belongs to.

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest = True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
#number of vechicles in each bin
df['horsepower-binned'].value_counts()
```

```python
#plot the distribution of each bin

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df['horsepower-binned'].value_counts())

#set x/y labels and the title of the plot
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('counts')
plt.pyplot.title('horsepower bins')

plt.pyplot.show()
```

### Indicator variable

#### What is indicator variable?

An indicator variable or a dummy variable is a numerical variable used to label categories. they are called 'dummies' because the numbers themselves don't have inherent meaning.

#### Why use indicator variables?

We use indicator variables so that we can use categorical variables for regression analysis in the later modules. 

#### Example 
The column 'fuel-type' has 2 unique values - 'gas' and 'diesel'. Regression doesn't understand words, only numbers. To use this attribute in reggresion analysis, you can convert 'fuel-type' to indicator variables.

Use the pandas 'get_dummies' to assign nummerical values to different categories of fuel type.

```python
df.columns
```

```python
# get the indicator variables and assign it to data frame 'dummy_variables_1'
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
#merge data frame df and dummy_variable_1
df = pd.concat([df,dummy_variable_1], axis = 1 )

#drop original column 'fuel-type' from from df
df.drop('fuel-type', axis = 1, inplace = True)
df.head()
```

```python
df['aspiration'].value_counts()
```

```python
dummy_variable_2 = pd.get_dummies(df['aspiration'])
df = pd.concat([df, dummy_variable_2], axis = 1)
df.drop('aspiration', axis = 1, inplace = True)
df.head()
```
